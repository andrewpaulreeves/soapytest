```python
import numpy
%matplotlib nbagg
from matplotlib import pyplot
from tqdm import tnrange
import pandas
```

```python
import soapy
import aotools
```

```python
CONFIGFILE = "conf/simulation/isoplanaticerror.yaml"
```

```python
LAYER_ALTITUDES = [0, 5000, 10000, 15000]
GS_ANGLES = [0, 2, 4, 6, 8, 10, 15, 20,]
RUNS = 4
TEL_DIAM = 8.
R0 = 0.20
WAVELENGTH = 500e-9
```

```python
data_frame = pandas.DataFrame()

for r in tnrange(RUNS):
    for nl in tnrange(len(LAYER_ALTITUDES)):
        for na in range(len(GS_ANGLES)):

            print('\n***\nSet Layer to {}m, GS angle to {}"\n***'.format(LAYER_ALTITUDES[nl], GS_ANGLES[na]))
            sim = soapy.Sim(CONFIGFILE)
            sim.config.atmos.scrnHeights[0] = LAYER_ALTITUDES[nl]
            sim.config.wfss[0].position[0] = GS_ANGLES[na]
            sim.aoinit()
            sim.makeIMat()
            sim.aoloop()

            data_series = pandas.Series(dict(sim.config.sim.saveHeader))
            data_series["GS_position"] = GS_ANGLES[na]
            data_series["layer_altitudes"] = LAYER_ALTITUDES[nl]
            data_series["long_strehl"] = sim.longStrehl[0][-1]
            data_series["wavefront_error"] = sim.WFE[0][-1]

            data_frame = data_frame.append(data_series, ignore_index=True)
```

```python
data_frame
```

```python
fig = pyplot.figure()
for nl in LAYER_ALTITUDES:
    df = data_frame[data_frame.layer_altitudes==nl]
    pyplot.scatter(df.GS_position, df.long_strehl, label="Layer Altitude: {}m".format(nl))
    pyplot.legend()
    pyplot.xlabel("Off-Axis Guide Star Position (arc-secs)")
    pyplot.ylabel("Strehl Ration @ 1.65 $\mu$m")
```
